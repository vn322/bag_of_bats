# Вероятность

**Вероятность** — это мера уверенности в том, что событие произойдет, часто выражаемая в процентах. Вот некоторые вопросы, для которых может потребоваться оценка вероятности:

*   Насколько вероятно, что я получу 7 решек при 10
подбрасываниях честной монеты?
*   Каковы шансы этого кандидата выиграть на выборах?
*   Опоздает ли мой рейс?
*   Насколько я уверен, что продукт бракованный?

<details>
  <summary>Текст</summary>

  Наиболее распространенный способ выражения вероятности — это процентное соотношение, например: «Существует вероятность **70%**, что мой рейс будет задержан». Мы будем обозначать эту вероятность как $P(X)$, где $X$ — это интересующее нас событие. Однако, при работе с вероятностями чаще используется десятичное выражение (в данном случае **0.7**), которое должно находиться в диапазоне от **0.0** до **1.0**.

  $$
  P(X) = 0.70
  $$

</details>

- **Вероятность:** используется для предсказания будущих событий, основанных на известных параметрах или моделях.

- **Правдоподобие:** используется для нахождения параметров моделей, исходя из наблюдаемых данных.

<details>
  <summary>Текст</summary>

  Понятие правдоподобия (*aнг. likelihood*) близко по смыслу к понятию вероятность, и их легко спутать. Однако, нам следует четко разграничить эти понятия.

  Вероятность даёт количественную оценку предсказаний событий, которые еще не произошли, тогда как правдоподобие измеряет частоту событий, которые уже произошли. В статистике и машинном обучении мы часто используем правдоподобие (**прошлое**) в виде данных для того, чтобы сделать модель, которая будет выдавать вероятности (**будущее**).

</details>

Важно отметить, что вероятность события должна быть строго между **0%** и **100%**, или между **0** и **1**. Логически это означает, что вероятность того, что событие не произойдет, вычисляется вычитанием вероятности события из **1**:

$$
P(X) = 0.70
$$

$$
P(\bar{X}) = 1.00 - 0.70 = 0.30
$$

Иногда, вероятность может быть выражена в виде шансов (odds), например, **7:3**, **7/3** или **2.(3)**. Чтобы преобразовать шансы $O(X)$ в вероятность $P(X)$, используйте следующую формулу:

$$
P(X) = \frac{O(X)}{1 + O(X)}
$$

<details>
  <summary>Текст</summary>

  Да, если у вас есть шанс 7 к 3, вы можете преобразовать их в пропорциональную вероятность следующим образом:

  $$
  P(X) = \frac{O(X)}{1 + O(X)}
  $$

  $$
  P(X) = \frac{7/3}{1 + 7/3}
  $$

  $$
  P(X) = 0.7
  $$

  Для того чтобы обратно преобразовать вероятности в шансы нужно просто разделить вероятность события на вероятность того, что это событие не произойдет. Формула выглядит следующим образом:

  $$
  O(X) = \frac{P(X)}{1-P(X)}
  $$

  $$
  O(X) = \frac{0.70}{1-0.70}
  $$

  $$
  O(X) = 7 / 3
  $$
</details>

# В чем отличие между теорией вероятности и статистикой?

Порой термины **"теория вероятности"** и **"статистика"** используется взаимозаменяемо, и хотя путаница между этими дисциплинами понятна, они имеют различия. **Теория вероятности** — это чисто теоретический расчет того, насколько вероятно событие, это вычисление не требует данных. **Статистика**, с другой стороны, не имеет смысла без данных и использует их для расчета вероятностей, а также предоставляет инструменты для описания данных.

<details>
  <summary>Текст</summary>

  Рассмотрим пример предсказания вероятности выпадения **4** на игральной кости. Подходя к этой проблеме с точки зрения чистой теории вероятности, мы просто говорим, что на кубике шесть граней. Мы предполагаем, что каждая грань может выпасть одинаково вероятно, поэтому вероятность выпадения **4** составляет **1/6**.

  Однако ревностный статистик может сказать: «Нет! Нам нужно бросить кубик, чтобы получить данные. Если мы сможем сделать **30** бросков или больше, и чем больше бросков мы совершим, тем лучше, только тогда у нас будут данные, чтобы определить вероятность выпадения **4**». Этот подход может показаться нелепым, если мы предполагаем, что кубик честный, но что, если он не таков? В этом случае сбор данных — единственный способ определить вероятность выпадения **4**. Мы поговорим о проверке гипотез в следующей лекции.

  Еще можно воспринимать это отличие между теорвером и статиситкой в том, что теорвер по известным теоретическим распределениям считает вероятности реализаций случайных величин, а матстат наоборот - по реализациям пытается рассчитать распределения и оценки.
</details>

# Математика вероятностей

Когда мы имеем дело с вероятностью отдельного события
$P(X)$, известной как частная вероятность, то работать с вероятность достаточно просто. Но когда мы начинаем комбинировать вероятности различных событий, то нам уже нужны математические операции над вероятностями.

### Частотное определение вероятности

**Частотное** определение вероятности основывается на идее частоты возникновения события в длинной серии повторяющихся экспериментов. Согласно этому подходу, вероятность события $A$ определяется как предел отношения числа случаев наступления события $A$ к общему числу проведенных экспериментов, при стремлении числа экспериментов к бесконечности.

<details>
  <summary>Текст</summary>

  #### Формальное определение
  Если эксперимент повторяется множество раз, и $n$ — общее число экспериментов, а $n_A$ — количество случаев, когда наступило событие $A$, то вероятность $P(A)$ можно определить как:

  $$
  P(A) = \lim_{n \to \infty} \frac{n_A}{n}
  $$

  #### Пример
  Рассмотрим подбрасывание монеты. Если мы подбросили монету **1000** раз и видим, что орел выпал **485** раз, то частотная оценка вероятности выпадения орла будет:

  $$
  P(\text{ОРЕЛ}) = \frac{485}{1000} = 0.485
  $$

</details>

### Геометрическое определение вероятности

**Геометрическое** определение вероятности используется в ситуациях, когда эксперимент включает пространство возможных исходов, которые имеют геометрическую структуру, такую как длина, площадь или объем. Согласно этому подходу, вероятность события определяется как отношение меры множества, соответствующего событию, к мере всего множества исходов.

<details>
  <summary>Текст</summary>

#### Формальное определение
Если событие $A$ занимает часть геометрического пространства, например, участок длины, площади или объема, то вероятность $P(A)$ определяется как отношение меры этого участка к мере всего пространства.

$$
P(A) = \frac{\text{Мера множества } A}{\text{Мера всего пространства}}
$$

#### Пример
Предположим, что мы случайным образом выбираем точку на отрезке длиной **10 см**. Пусть событие $A$ состоит в том, что выбранная точка попадает на отрезок длиной **4 см**. Тогда вероятность этого события можно определить как:

$$
P(A) = \frac{4 \text{ см}}{10 \text{ см}} = 0.4
$$

</details>

# Правило умножения

Предположим, у вас есть честная монета и честный шестигранный кубик. Вы хотите найти вероятность выпадения орла на монете и шестерки на кубике. Это две отдельные вероятности двух отдельных событий, но мы хотим найти вероятность того, что оба события произойдут вместе. Это называется **совместной вероятностью.**

Представьте себе совместную вероятность как оператор **И** (*aнгл.* **AND**). Мы хотим найти вероятность того, что выпадет орел И выпадет шестерка. Мы хотим, чтобы оба события произошли вместе, так как же мы можем рассчитать эту вероятность?

У монеты две стороны, а у кубика шесть граней, поэтому вероятность выпадения орла составляет **1/2**, а вероятность выпадения шестерки — **1/6**. Вероятность того, что оба события произойдут (предполагая, что они независимы, об этом позже!), просто равна произведению двух вероятностей:

$$
P(\text{ОРЕЛ И 6}) = P(\text{ОРЕЛ}) \times P(\text{6})$$

Таким образом:

$$
P(\text{ОРЕЛ И 6}) = \frac{1}{2} \times \frac{1}{6} = \frac{1}{12}
$$

То есть вероятность совместного события — выпадения орла на монете и шестерки на кубике — составляет **1/12** или примерно **8.(33333)%**.

<details>
  <summary>Текст</summary>

  Выглядит достаточно просто, но почему это так? В большинстве случаев вероятность может быть вычислена, путем рассмотрения всех возможных комбинаций событий, в этом вопросе нам часто помогает раздел математики называемой комбинаторикой. Сгенерируем вообще возможные исходы для монеты и кубика, сочетания орла **О** и решки **Р** с числами от **1** до **6**.

  Обратите внимание, что звездочкой («*») отмечен интересующий нас исход, где выпадает орел и шестерка:

  $O1\quad O2\quad O3\quad O4\quad O5\quad O6^{*}\quad P1\quad P2\quad P3\quad P4\quad P5\quad P6$

  При подбрасывании нашей монеты и бросании кубика существует **12** возможных исходов. Единственный исход, который нас интересует, это **О6**, то есть выпадение орла и шестерки. Поскольку существует только один исход, который удовлетворяет нашему условию, и всего **12** возможных равновероятных исходов, вероятность выпадения орла и шестерки составляет **1/12**.

</details>

# Правило сумм

Мы обсудили совместные вероятности, которые представляют собой вероятность одновременного наступления двух или более событий. Но что насчет вероятности наступления события **A** или **B**? Когда мы имеем дело с операциями **ИЛИ** (*англ.* **OR**)?

<details>
  <summary>Текст</summary>

  Начнем с событий, которые являются **взаимно исключающими**, то есть событий, которые не могут произойти одновременно. Например, если я бросаю один кубик, я не могу одновременно получить **4** и **6**. Я могу получить только один исход. Вычислить вероятность объединения для таких случаев просто. Я просто складываю их. Если я хочу получить вероятность выпада **4** или **6** при бросании кубика, то это будет $2/6 = 1/3$:

  $$
  P(4) = \frac{1}{6}
  $$

  $$
  P(6) = \frac{1}{6}
  $$

  $$
  P(4 \ OR \ 6) = P(4) + P(6) = \frac{1}{6} + \frac{1}{6} = \frac{1}{3}
  $$
</details>

<details>
  <summary>Текст</summary>

  Но что насчет не взаимно исключающих событий, то есть событий, которые могут произойти одновременно? Вернемся к примеру с подбрасыванием монеты и броском кубика. Какова вероятность выпадения ОРЛА **ИЛИ** ШЕСТРЕКИ? Прежде чем вы захотите сложить эти вероятности, давайте снова сгенерируем все возможные исходы и выделим те, которые нас интересуют:

  $$ O1^{*} \ O2^{*} \ O3^{*} \ O4^{*} \ O5^{*} \ O6^{*} \ P1\ P2\ P3\ P4\ P5\ P6^{*} \
  $$

  Здесь нас интересуют все исходы с орлом, а также исходы с шестеркой. То есть нам подходит 7 из 12 исходов, которые нас интересуют, то есть **7/12** и есть вероятность, которую мы ищем.

  Давайте попробуем понять какое правило работает в данном случае. Что произойдёт, если сложить вероятности выпадения ОРЛА и ШЕСТЕРКИ? Мы получим другое (и неверное!) значение **2/3**:

  $$
  P(\text{ОРЕЛ}) = \frac{1}{2}
  $$
  $$
  P(\text{6}) = \frac{1}{6}
  $$
  $$
  P(\text{ОРЕЛ ИЛИ 6}) = \frac{1}{2} + \frac{1}{6} = \frac{4}{6} = \frac{2}{3}
  $$

  Почему так? Посмотрим комбинации исходов подбрасывания монеты и броска кубика еще раз и и в них можно найти что-то подозрительное. Обратите внимание, что при сложении вероятностей мы дважды учитываем вероятность выпадения шестерки как в **O6**, так и в **P6**! Доведем это до абсурад и найдем вероятность выпадения орла или выпадения числа от 1 до 5:

  $$
  P(\text{ОРЕЛ}) = \frac{1}{2}
  $$
  $$
  P(\text{1 ДО 5}) = \frac{1}{6} \times 5 = \frac{5}{6}
  $$

  $$
   P(\text{ОРЕЛ ИЛИ 1 ДО 5}) = \frac{1}{2} + \frac{5}{6} = \frac{1}{2} + \frac{5}{6} = \frac{8}{6} = 1.(3)
   $$

  Здесь мы вообще получили вероятность **133.333%**, что явно неверно, поскольку вероятность не может быть выше **100%** или **1**. Проблема в том, что мы дважды учитываем некоторые исходы.
</details>

Если внимательно посмотреть и долго, то можно понять, что логическим способом устранения двойного учета в объединенной вероятности является вычитание совместной вероятности. Это называется правилом суммы вероятностей и гарантирует, что каждое совместное событие учитывается только один раз:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

Возвращаясь к нашему примеру вычисления вероятности выпадения орла или шестерки, нужно вычесть совместную вероятность выпадения орла и шестерки из объединенной вероятности:

$$
P(\text{ОРЕЛ}) = \frac{1}{2}
$$

$$
P(\text{6}) = \frac{1}{6}
$$

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

$$
P(\text{ОРЕЛ ИЛИ 6}) = \frac{1}{2} + \frac{1}{6} - \left(\frac{1}{2} \times \frac{1}{6}\right) = \frac{7}{12}
$$

Итак, правильная вероятность выпадения орла или шестерки составляет **7/12**.

<details>
  <summary>Текст</summary>

  Обратите внимание, что эта формула также применяется к взаимно исключающим событиям. Если события являются взаимно исключающими, то есть допускается только один из исходов $A$ или $B$, но не оба, тогда совместная вероятность $ P(A \cap B) $ будет равна нулю и, следовательно, ее не будет в формуле. В этом случае остаётся просто сложить вероятности событий, как мы сделали ранее.

  В итоге, когда у нас есть объединенная вероятность двух или более событий, которые не являются взаимно исключающими, обязательно надо вычитать совместную вероятность, чтобы никакие вероятности не были учтены дважды.
</details>

### Условная вероятность и теорема Байеса
Тема в теории вероятности, которая сильно путает людей, — это концепция условной вероятности, то есть вероятность наступления события $A$ при условии, что событие $B$ уже произошло. Это обычно записывается как $ P(A \mid B) $.

Предположим, исследование утверждает, что 85% больных раком пили кофе. Как вы реагируете на это утверждение? Заставляет ли это вас беспокоиться и желать отказаться от вашего любимого утреннего напитка? Давайте сначала определим это как условную вероятность $ P(\text{КОФЕ} \mid \text{РАК}) $. Это вероятность того, что человек пьет кофе, при условии, что у него рак.

Давайте сравним это с процентом людей, у которых диагностирован рак (0.5% согласно cancer.gov), и процентом людей, которые пьют кофе (65% согласно statista.com) в США:

$$
P(\text{РАК}) = 0.005
$$

$$
P(\text{КОФЕ}) = 0.65
$$

$$ P(\text{РАК} | \text{КОФЕ}) = 0.85
$$


Посмотрите на эти числа и задумайтесь, действительно ли проблема в кофе. Заметьте снова, что только у **0.5%** населения есть рак в любой данный момент времени. Однако **65%** населения регулярно пьют кофе. Если кофе способствует развитию рака, разве мы не должны видеть намного более высокие показатели рака, чем **0.5%**? Разве это число не должно быть ближе к **65%**?

Это коварная особенность процентов. Они могут казаться значительными без всякого контекста, и заголовки в СМИ могут это эксплуатировать для привлечения внимания: «Новое исследование показывает, что **85%** больных раком пьют кофе», — можно прочитать. Конечно, это нелепо, потому что мы взяли распространенный признак (употребление кофе) и связали его с редким признаком (наличие рака).

Причина, по которой люди так легко могут запутаться в условных вероятностях, заключается в том, что направление условия имеет значение, и два условия ошибочно принимаются за равные. "Вероятность иметь рак, если вы пьете кофе"  отличается от «вероятности того, что вы пьете кофе, если у вас есть рак». Проще говоря: мало кто из пьющих кофе имеет рак, но многие больные раком пьют кофе.

Если нас интересует вопрос, способствует ли кофе развитию рака, то нам нужна условная вероятность "иметь рак, если вы пьете кофе".

$$
P(\text{КОФЕ}|\text{РАК}) = 0.85
$$

$$
P(\text{РАК}|\text{КОФЕ}) = \text{?}
$$


Как но как нам поменять условия и событие местами? Для этого существует **теоремой Байеса**, и мы можем использовать её для того, чтобы поменять условие и событие местами:

$$
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
$$

$P(A \mid B)$ — условная вероятность события $A$ при условии, что произошло событие $B$, $P(B \mid A)$ — условная вероятность события $B$ при условии, что произошло событие $A$, $P(A)$ — вероятность события $A$, $P(B)$ — вероятность события $B$.

Если мы подставим имеющуюся информацию в эту формулу, мы сможем решить задачу нахождения вероятности того, что у кого-то будет рак, если он пьет кофе:


$$
P(\text{РАК} \mid \text{КОФЕ}) = \frac{P(\text{КОФЕ} \mid \text{РАК}) \cdot P(\text{РАК})}{P(\text{КОФЕ})}
$$

$$
P(\text{РАК} \mid \text{КОФЕ}) = \frac{0.85 \times 0.005}{0.65}
$$

$$
P(\text{Рак} \mid \text{Кофе}) = \frac{0.00425}{0.65} \approx 0.00654
$$

Таким образом, вероятность того, что у кого-то будет рак, если он пьет кофе, приближенно равна **0.654%**.

```python
# Пример 1
p_coffee_drinker = 0.65
p_cancer = 0.005
p_coffee_drinker_given_cancer = 0.85
p_cancer_given_coffee_drinker = p_coffee_drinker_given_cancer * p_cancer / p_coffee_drinker

print(p_cancer_given_coffee_drinker)
```

Таким образом, вероятность того, что у кого-то будет рак, если он пьет кофе, составляет всего **0.65%**! Это число сильно отличается от вероятности того, что кто-то пьет кофе, если у него есть рак, которая составляет **85%**. Теперь вы видите, почему место условия и события имеет значение? **Теорема Байеса** может быть использована для последовательного учета нескольких условных вероятностей, чтобы постоянно обновлять наши убеждения на основе новой информации.

Давайте снова рассмотрим совместные вероятности и как они взаимодействуют с условными вероятностями. Я хочу найти вероятность того, что кто-то пьет кофе И у него есть рак. Должен ли я умножать $ P(\text{КОФЕ}) $ и $ P(\text{РАК}) $? Или я должен использовать $ P(\text{КОФЕ} \mid \text{РАК}) $ вместо $ P(\text{КОФЕ}) $, если это значение доступно? Какой вариант использовать?

**Вариант 1:**

$ P(\text{КОФЕ}) \times P(\text{РАК}) = 0.65 \times 0.005 = 0.00325 $

**Вариант 2:**

$ P(\text{КОФЕ} \mid \text{РАК}) \times P(\text{РАК}) = 0.85 \times 0.005 = 0.00425 $

Если уже установлено, что наша вероятность относится только к людям с раком, логично использовать $ P(\text{КОФЕ} \mid \text{РАК}) $ вместо $ P(\text{КОФЕ}) $? Одно значение более конкретное и относится к уже установленному состоянию. Поэтому следует использовать $ P(\text{КОФЕ} \mid \text{РАК}) $, так как $ P(\text{РАК}) $ уже является частью нашей совместной вероятности. Это означает, что вероятность того, что у кого-то будет рак и он будет пить кофе, составляет **0.425%**:

$$
 P(\text{КОФЕ И РАК}) = P(\text{КОФЕ} \mid \text{РАК}) \times P(\text{РАК}) = 0.85 \times 0.005 = 0.00425
 $$

Эта совместная вероятность также применяется в другом направлении. Я могу найти вероятность того, что кто-то пьет кофе и у него есть рак, умножая $P(\text{РАК} \mid \text{КОФЕ}) $ и $ P(\text{КОФЕ}) $. Как можно заметить, я прихожу к тому же результату:

$$
P(\text{РАК} \mid \text{КОФЕ}) \times P(\text{КОФЕ}) = 0.0065 \times 0.65 = 0.00425
$$

Если бы у нас не было никаких условных вероятностей, тогда максимум, что мы можем сделать, это умножить $ P(\text{КОФЕ}) $ и $ P(\text{РАК}) $, как показано здесь:

$$
P(\text{КОФЕ}) \times P(\text{РАК}) = 0.65 \times 0.005 = 0.00325
$$

Теперь подумайте об этом: если событие $A$ не влияет на событие $B$, что это значит для условной вероятности $ P(B \mid A) $? Это означает, что $ P(B \mid A) = P(B) $, то есть событие $A$ не влияет на вероятность наступления события $B$. Таким образом, мы можем обновить нашу формулу совместной вероятности, независимо от того, зависимы ли два события, следующим образом:

$$
P(A \text{ И } B) = P(B) \times P(A \mid B)
$$

И наконец, поговорим о объединениях и условных вероятностях. Если я хочу вычислить вероятность наступления $A$ или $B$, но $A$ может влиять на вероятность $B$, мы обновляем наше правило суммы следующим образом:

$ P(A \text{ ИЛИ } B) = P(A) + P(B) - P(A \mid B) \times P(B) $

Напоминаю, что это также применяется к взаимно исключающим событиям. Правило суммы $ P(A \mid B) \times P(B) $ даст **0**, если события $A$ и $B$ не могут произойти одновременно.

# Биномиальное распредление

Давайте рассмотрим пример из реальной жизни.

Предположим, что вы работаете над новым реактивным двигателем и провели **10** испытаний. Результаты дали восемь успешных и два неудачных исхода:

# ✓ ✓ ✓ ✓ ✓ ✘ ✓ ✘ ✓ ✓

Вы надеялись получить **90%** успешности, но на основе этих данных вы заключаете, что ваши испытания провалились, достигнув только **80%** успешности. Каждое испытание занимает много времени и дорого обходится, поэтому вы решаете вернуться к чертежам и переработать конструкцию.

Однако один из ваших инженеров настаивает на проведении большего количества испытаний. «Единственный способ узнать наверняка — это провести больше испытаний», — утверждает он. «Что, если большее количество испытаний принесет **90%** или более успехов? Ведь если подбросить монету 10 раз и получить **8** орлов, это не значит, что монета "жульничает" на **80%**».

Вы коротко обдумываете аргумент инженера и понимаете, что он прав. Даже бросок "честной" монеты не всегда имеет равномерный результат, особенно при только **10** подбрасываниях. Вы, скорее всего, получите пять орлов, но также можете получить три, четыре, шесть или семь орлов. Вы можете даже получить **10** орлов, хотя это крайне маловероятно. Так как же определить вероятность **80%** успеха при предположении, что изначальная вероятность составляет **90%**?

Один из инструментов, который может быть полезен в данном случае, — это **биномиальное распределение**, которое измеряет, насколько вероятны $k$ успехов из $n$ испытаний при данной вероятности $p$:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

где:

$$
\binom{n}{k} = \frac{n!}{k!(n-k)!}
$$

- $ n $ — число испытаний.
- $ k $ — число успехов.
- $ p $ — вероятность успеха в одном испытании.

Визуально биномиальное распределение выглядит как на *рисунке 2*.

Здесь мы видим вероятность $k$ успехов для каждой строки из 10 испытаний. Это биномиальное распределение предполагает вероятность $p$ в **90%**, что означает, что шанс на успех составляет **0.90** (или **90%**). Если это верно, то вероятность получить **8** успехов из **10** испытаний составляет 0.1937. Вероятность получить **1** успех из **10** испытаний крайне мала — **0.000000008999**, поэтому эта строка даже не видна.



```python
#@title <b> Рисунок </b>
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Параметры биномиального распределения
n = 10  # Количество экспериментов
p = 0.9  # Вероятность успеха

# Значения возможных исходов (от 0 до n)
x = np.arange(0, n + 1)

# Расчет биномиального распределения для каждого значения x
y = binom.pmf(x, n, p)

# Построение графика
plt.bar(x, y, color='blue')
plt.xlabel('Количество успехов')
plt.ylabel('Вероятность')
plt.title(f'Биномиальное распределение (n={n}, p={p})')
plt.show()
```

Мы также можем вычислить вероятность восьми или менее успехов, сложив вероятности для восьми или менее успехов. Это даст нам вероятность **0.2639**.

Для того, чтобы использовать биноамиальное распределение мы можем подключить специальную библиотеку, такую как SciPy. *Пример 2* показывает, как мы используем функцию binom.pmf() из библиотеки SciPy для вывода всех **11** вероятностей для нашего биномиального распределения от **0** до **10** успехов.

```python
# Пример 2. Биномиальное распределение
from scipy.stats import binom

# n - количество испытаний
n = 10
# p - вероятность успеха в одном испытании
p = 0.8

# Вывод вероятностей для каждого возможного числа успехов от 0 до 10
for k in range(n + 1):
    probability = binom.pmf(k, n, p)
    print(f'Вероятность {k} успехов из {n} испытаний: {probability:.10f}')
```

Как видите, мы задаем $n$ как количество испытаний, $p$ как вероятность успеха для каждого испытания и $k$ как количество успехов, для которых мы хотим найти вероятность. Мы итерируемся по каждому числу успехов $x$ с соответствующей вероятностью, которую бы мы увидели для такого числа успехов. Как видно из вывода, наиболее вероятное число успехов — девять.

Но если мы сложим вероятность для восьми или менее успехов, то получим **0.2639**. То есть существует **26.39%** вероятность того, что мы увидим восемь или менее успехов, даже если базовая частота успехов составляет **90%**. Поэтому инженер может быть и прав: вероятность **26.39%** - это достаточно большая вероятность, чтобы еще надеяться, что испытания прошли успешно 😀

Однако мы сделали предположение в нашей модели, о котором мы дальше поговорим, обсуждая бета-распределение.

#Бета-распредление
Что именно мы предположили в своей модели испытаний двигателя, используя биномиальное распределение?

У нас есть базовое предположение, что базовая вероятность успеха **90%**. Это не означает, что модель бесполезна. Это значит, что мы показали, что при вероятности успеха в **90%** существует вероятность **26.39%** увидеть **8** или менее успехов из **10** испытаний. Поэтому инженер, безусловно, не ошибается, говоря, что вероятность успеха может быть и **90%**.

А что будет если мы рассмотрим другие варианты вероятности успеха. Они тоже могут привести к наблюдаемому нами результату **8** успехов и **10** испытаний, помимо вероятности **90%**? Можем ли мы увидеть **8** из **10** успехов при вероятности успеха в **80%**? **70%**? **30%**? Когда мы фиксируем **8** успехов из **10** испытаний, можем ли мы расчитатать вероятности для различных вероятностей успехов? Да именно так вероятность вероятности 😀
У меня уже кружится голова.

Мы конечно можем рассмотреть различные варианты вероятности успехов и найти тот, который вероятнее всего подходит под наше наблюдение 8 успехов из 10 испытаний. Однако, вместо того чтобы создавать бесконечное количество биномиальных распределений для ответа на этот вопрос, существует аналитический подход, который называется **Бета-распределение**. Она позволяет нам увидеть вероятность различных вероятностей успеха для наступления события (нашего наблюдения $\alpha$ успехов из $\beta$ испытаний).



```python
#@title <b> Рисунок </b>

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta

# Параметры бета-распределения
alpha = 9
beta_ = 3

# Создаем массив значений для оси x от 0 до 1
x = np.linspace(0, 1, 1000)

# Вычисляем плотность вероятности для каждого значения x
y = beta.pdf(x, alpha, beta_)

# Построение графика
plt.plot(x, y, 'r-', label=f'α={alpha}, β={beta_}')
plt.xlabel('Скрытая вероятность успеха')
plt.ylabel('Правдоподобие 8/10 успехов')
plt.title('Бета-распределение')
#plt.legend()
plt.grid()
plt.show()
```

Рисунок 4. Бета распределение

Ось **x** представляет все возможные вероятности успеха от **0.0** до **1.0** (от **0%** до **100%**), а ось y представляет распределение вероятности этих вероятностей при условии наблюдения **8** успехов и **2** неудач. Другими словами, бета-распределение позволяет нам увидеть вероятности различных базовых вероятностей, учитывая **8/10** успехов. Подумайте об этом как о мета-вероятности!

Также обратите внимание, что бета-распределение является непрерывной функцией в отличие от дискретных целых чисел в биномиальном распределении. Это несколько усложнит математические расчеты с бета-распределением, так как конкретное значение распределение вероятности по оси y не является вероятностью. Вместо этого мы находим вероятности интегрирую распределение вероятности.

**Бета-распределение** — это распеделение вероятностей, что означает, что площадь под всей кривой равна **1.0**. Чтобы найти вероятность, нам нужно найти площадь в заданном диапазоне. Например, если мы хотим оценить вероятность того, что **8** успехов из **10** экспериментов приведут к успеху **90%** или выше, нам нужно найти площадь между **0.9** и **1.0**, что составляет **0.225**, как показано на *рисунке 5*.




```python
#@title <b> Рисунок </b>

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta

# Параметры бета-распределения на основе числа успехов и неудач
successes = 8
trials = 10
alpha = successes + 1
beta_ = trials - successes + 1

# Создаем массив значений для оси x от 0 до 1
x = np.linspace(0, 1, 1000)

# Вычисляем плотность вероятности для каждого значения x
y = beta.pdf(x, alpha, beta_)

# Построение графика
plt.plot(x, y, 'r-', label=f'α={alpha}, β={beta_}')
plt.fill_between(x, y, where=(x >= 0.9) & (x <= 1), color='red', alpha=0.3)
plt.xlabel('Скрытая вероятность успеха')
plt.ylabel('Правдоподобие 8/10 успехов')
plt.title('Бета-распределение (8 успехов из 10 испытаний)')
#plt.legend()
plt.grid()
plt.show()
```

Рисунок 5. Площадь между базовой вероятностью успеха 0.9 и 1.0, которая равна **0.225**


Как и в случае с биномиальным распределением, мы можем использовать библиотеку SciPy для реализации бета-распределения. С каждым непрерывным распределением в паре идет функция кумулятивной плотности (CDF - Cumulative Distribution Function), которая вычисляет площадь до заданного значения $x$. Допустим, я хочу вычислить площадь до 90% (от 0.0 до 0.90), как показано на *рисунке 6*.


```python
#@title <b> Рисунок </b>

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta

# Параметры бета-распределения на основе числа успехов и неудач
successes = 8
trials = 10
alpha = successes + 1
beta_ = trials - successes + 1

# Создаем массив значений для оси x от 0 до 1
x = np.linspace(0, 1, 1000)

# Вычисляем плотность вероятности для каждого значения x
y = beta.pdf(x, alpha, beta_)

# Построение графика
plt.plot(x, y, 'r-', label=f'α={alpha}, β={beta_}')
plt.fill_between(x, y, where=(x <= 0.9), color='red', alpha=0.3)
plt.xlabel('Скрытая вероятность успеха')
plt.ylabel('Правдоподобие 8/10 успехов')
plt.title('Бета-распределение (8 успехов из 10 испытаний)')
#plt.legend()
plt.grid()
plt.show()
```

Рисунок 6. Расчет площади под кривой вплоть до 0.9.

Достаточно просто можно использовать SciPy с функцией beta.cdf(), чтобы расчитать эту площадь под кривой. Единственные параметры, которые нам нужно предоставить этой фукнции, это значение $x$, количество успехов $a$ и количество неудач $b$ (в *примере 3*).

```python
# Пример 3
from scipy.stats import beta
a = 8
b = 2
p = beta.cdf(.90, a, b)
print(p)
```

Итак, согласно нашему расчету, вероятность того, что базовая вероятность успеха составляет **90%** или меньше, равна **77.48%**.
Как нам вычислить вероятность того, что успех будет **90%** или больше, как показано на *рисунке 7*?


Наша функция **CDF** вычисляет площадь только слева от нашей границы, а не справа. Подумайте о наших правилах вероятности: для распределения вероятностей общая площадь под кривой равна **1.0**. Если мы хотим найти обратную вероятность события (больше **0.90** вместо меньше **0.90**), мы просто вычитаем вероятность быть меньше **0.90** из **1.0**, и оставшаяся вероятность будет соответствовать значениям больше **0.90**. *Рисунок 7* иллюстрирует, как мы выполняем это вычитание.

Рисунок 7. Пойск вероятности, того, что вероятность успеха больше 90%.


```python
# Пример 4
from scipy.stats import beta
a = 8
b = 2
p = 1.0 - beta.cdf(.90, a, b)
# 0.22515902199999993
print(p)
```

Это означает, что при **8** успешных из **10** испытаниях двигателя вероятность того, что базовая вероятность успехов составляет **90%** или больше, составляет всего **22.5%**. Но существует около **77.5%** вероятности, что она меньше **90%**. Шансы не в нашу пользу, что наши тесты были успешными, но мы могли бы рискнуть этой **22.5%** вероятностью, проведя больше тестов. Если наш финансовый директор разрешит финансирование для **26** дополнительных испытаний, в результате которых будет **30** успехов и **6** неудач, наше бета-распределение будет выглядеть как на *рисунке 8*.

```python
#@title <b> Рисунок </b>

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta

# Параметры бета-распределения на основе числа успехов и неудач
successes = 30
trials = 36
alpha = successes + 1
beta_ = trials - successes + 1

# Создаем массив значений для оси x от 0 до 1
x = np.linspace(0, 1, 1000)

# Вычисляем плотность вероятности для каждого значения x
y = beta.pdf(x, alpha, beta_)

# Построение графика
plt.plot(x, y, 'r-', label=f'α={alpha}, β={beta_}')
plt.fill_between(x, y, where=(x >= 0.9) & (x <= 1), color='red', alpha=0.3)
plt.xlabel('Скрытая вероятность успеха')
plt.ylabel('Правдоподобие 30/36 успехов')
plt.title('Бета-распределение (30 успехов из 36 испытаний)')
#plt.legend()
plt.grid()
plt.show()
```

Рисунок 8. Бета распредление после 30 успехов и 6 неудач.

Обратите внимание, что наше распределение стало более
узким, что свидетельствует о большей уверенности в том, что базовая верояность успехов находится в более узком диапазоне. К сожалению, наша вероятность достижения минимальной вероятности успехов в 90% уменьшилась, снизившись с 22.5% до 13.16% (пример 5).

```python
# Пример 5
from scipy.stats import beta
a = 30
b = 6
p = 1.0 - beta.cdf(.90, a, b)
# 0.13163577484183708
print(p)
```

На этом этапе может быть хорошей идеей прекратить тесты, если только вы не хотите продолжать рисковать ради той **13.16%**-ной вероятности и надеяться, что пик сместится вправо.
И последнее, но не менее важное, как бы нам рассчитать площадь в середине? Что, если я хочу найти вероятность того, что моя базовая частота успехов находится между **80%** и **90%**, как показано на *рисунке 9*?




```python
#@title <b> Рисунок </b>

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta

# Параметры бета-распределения на основе числа успехов и неудач
successes = 30
trials = 36
alpha = successes + 1
beta_ = trials - successes + 1

# Создаем массив значений для оси x от 0 до 1
x = np.linspace(0, 1, 1000)

# Вычисляем плотность вероятности для каждого значения x
y = beta.pdf(x, alpha, beta_)

# Построение графика
plt.plot(x, y, 'r-', label=f'α={alpha}, β={beta_}')
plt.fill_between(x, y, where=(x >= 0.8) & (x <= 0.9), color='red', alpha=0.3)
plt.xlabel('Скрытая вероятность успеха')
plt.ylabel('Правдоподобие 30/36 успехов')
plt.title('Бета-распределение (30 успехов из 36 испытаний)')
#plt.legend()
plt.grid()
plt.show()
```

Рисунок 9. Вероятность того, что базовая вероятность успеха лежит между 80% и 90%.

Подумайте внимательно, как вы могли бы подойти к этому вопросу. Что если мы вычтем площадь до **0.80** из площади до **0.90**? Даст ли это нам площадь между **0.80** и **0.90**? Да, и это даст площадь **0.3386** или **33.86%** вероятности. Вот как мы бы вычислили это на *Python* (*пример 6*).

```python
# Пример 6
from scipy.stats import beta
a = 8
b = 2
p = beta.cdf(.90, a, b) - beta.cdf(.80, a, b)
# 0.33863336200000016
print(p)
```

**Бета-распределение** — это инструмент для оценки вероятности наступления события или его отсутствия на основе ограниченного количества наблюдений. Оно позволяет нам анализировать вероятности вероятностей и обновлять наши оценки по мере поступления новых данных.
