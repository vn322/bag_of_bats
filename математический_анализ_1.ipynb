# Рациональные числа

# Приоритет арифметических операций

* Скобки
* Возведение в степень
* Умножение и деление
* Сложение и вычитание

Пример:

<br><br><br>
<br><br><br>
<br><br><br>

<details>
  <summary>Текст</summary>

  Мы надеемся, что вы уже знакомы с порядком выполнения **арифметических операций**, задающим последовательность приоритетов для каждого элемента математического выражения. Напоминаем: начинаем с вычисления выражений в **скобках**, затем переходим к **возведению в степень**, после чего следуют **умножение** и **деление**, а завершают список **сложение** и **вычитание**.

  Рассмотрим к примеру следующее выражение:

  $$
  2 × \frac{(3 + 2)^2}{5} - 4
  $$

  В начале мы вычисляем значение выражения в **скобках** $\textbf{(3 + 2)}$, получаем $\textbf{5}$:

  $$
  2 × \frac{(\mathbf{5})^2}{5} - 4
  $$

  После этого **возводим в степень**, в данном случае, возведем $\textbf{5}$ в квадрат. Получается $\textbf{25}$:

  $$
  2 × \frac{\mathbf{25}}{5} - 4
  $$

  Затем переходим к **умножению** и **делению**. Эти операции равнозначны, т.к. деление — это умножение на дробь. Произведем умножение $\textbf{2}$ на $\textbf{25/5}$, что равно $\textbf{50/5}$:

  $$
  \frac{\mathbf{50}}{5} - 4
  $$

  Следующим шагом является **деление**: $\textbf{50}$ делим на $\textbf{5}$, результатом будет $\textbf{10}$:

  $$
  \mathbf{10} - 4
  $$

  В заключение мы проводим операции **сложения** и **вычитания**. В данном случае, из $\textbf{10}$ вычитаем $\textbf{4}$, что равно $\textbf{6}$:


  $$
  10 - 4 = \mathbf{6}
  $$

  Если выполнить этот расчет в **Python**, результат будет равен $\textbf{6.0}$, что иллюстрируется в *примере 1*.

</details>

```python
# Пример 1: Приоритет арифметических  операций
my_value = 2 * (3 + 2)**2 / 5 - 4
print(my_value)
```

<details>
  <summary>Текст</summary>  
  
  Хотя это может казаться очевидным, в программировании важно использовать **скобки** даже если они избыточны. В коде верный результат возможен и без их применения, однако методичное использование **скобок** в сложных выражениях желательно для гарантии правильности **порядка операций**.

  Как пример, мы используем **скобки** для выделения части выражения с дробью, облегчая тем самым его восприятие и отделяя от остального контекста, как демонстрируется в *примере 2*.

</details>

```python
# Пример 2: Использование скобок для читаемости кода в Python
my_value = 2 * ((3 + 2)**2 / 5) - 4
print(my_value)
```

<details>
  <summary>Текст</summary>  
  
  Хотя обе версии кода корректны с технической точки зрения, последний вариант понятен лучше. В случае внесения изменений в код вами или кем-то другим, использование скобок облегчает понимание порядка выполнения операций. Такой подход служит защитой от нежелательных изменений в коде, помогая также избегать ошибок.

 # Функции

  * **Функция** — это правило, сопоставляющее каждому объекту из одной области ровно один объект из другой области, что позволяет переводить элементы из одного множества в элементы другого множества

  * Обозначение $ F: X \rightarrow Y $ указывает, что функция $\mathbf{F}$ принимает элементы из множества $\mathbf{X}$ и сопоставляет им элементы множества $\mathbf{Y}$, например, функция может преобразовывать вещественные числа $\mathbb{R}$ в комплексные числа $\mathbb{C}$

  * Линейная функция $\mathbf{y = 2x + 1}$ показывает, как для любого заданного значения $\textbf{x}$ можно найти соответствующее $\textbf{y}$

<details>
  <summary>Текст</summary>  

  **Функция** в математике — это правило, которое каждому объекту из одного набора (называемого областью определения) сопоставляет ровно один объект из другого набора (называемого областью значений). Это понятие можно представить как способ перевода или трансформации элементов из одного множества в элементы другого множества. **Функции** могут действовать между различными типами пространств, будь то числовые пространства, пространства функций, векторные пространства или даже более сложные структуры. Эта концепция является ключевой во многих областях математики и её приложениях, и вы будете встречаться с ней на протяжении всего курса.

  Обозначение ($F: X \rightarrow Y$) означает, что **функция** ($\mathbf{F}$) принимает элементы из множества ($\mathbf{X}$) (область определения) и сопоставляет им элементы множества ($\mathbf{Y}$) (область значений). Здесь ($\mathbf{X}$) может быть, например, множеством вещественных чисел ($\mathbb{R}$), а ($\mathbf{Y}$) — множеством комплексных чисел ($\mathbb{C}$), что означает, что **функция** ($\mathbf{F}$) преобразует вещественные числа в комплексные.

  Это универсальное определение отражает способность **функций** устанавливать связь между разнообразными объектами и областями, и является фундаментальной идеей, лежащей в основе многих математических теорий и методов.

  Говоря чуть проще, **функция** использует **входные переменные** (их еще называют **независимые переменные**), включает их в выражение, и в результате получается **выходная переменная** (известная также как **зависимая переменная**).
  В качестве примера рассмотрим следующую простую **линейную функцию**:

  $$
  y = 2x + 1
  $$

  Допустим, нам нужно взять числа с 1 по 5, каждое умножить на 2, а затем полученные значения сложить. Для описания этой процедуры мы можем использовать операцию суммирования.

  \sum_{i=1}^5 2 i = (2) \times 1 + (2) \times 2 + (2) \times 3 + (2) \times 4 + (2) \times 5 = 30

# Возведение в степень

* При возведении числа в степень основание умножается на себя определённое количество раз. Например, $ 2^3 = 2 \times 2 \times 2 = 8$. При умножении степеней с одинаковыми основаниями показатели степени складываются: $x^2 \times x^3 = x^{2+3} = x^5$.

* Отрицательный показатель степени является эквивалентом деления. Например, $x^{-3} = \frac{1}{x^3}$, что следует из сокращения одинаковых множителей в числителе и знаменателе: $\frac{x^2}{x^5} = x^{2-5} = x^{-3}$.

* Любое число, возведённое в степень 0, равно 1. Это объясняется свойством деления числа на само себя: $\frac{x^3}{x^3} = x^0 = 1$, исходя из чего $x^0 = 1$.


<details>
  <summary>Текст</summary>

  **Возведение в степень** - это процесс умножения числа на себя заданное количество раз. Возведение числа **2** в третью степень (обозначается как $\mathbf{2^3}$) подразумевает троекратное умножение двойки на себя:

  $$
  2^3 = 2 \times 2 \times 2 = 8
  $$

  *Основание* - это то число, которое возводится в степень, а показатель степени указывает, сколько раз основание умножается само на себя. В выражении $\mathbf{2^3}$ число **2** - это основание, а **3** - показатель степени.
  Рассмотрим умножение $\mathbf{x^2}$ на $\mathbf{x^3}$. Раскрытие степени через умножение и последующее сведение их к одному показателю степени выглядит так:

  $$
  x^2 \times x^3 = x \times x \times x \times x \times x = x^{(2 + 3)} = x^5
  $$

  При умножении чисел в степени с одинаковыми основаниями показатели степени складываются, чем и определяется правило произведения. Важно, что для применения этого правила основания должны совпадать.

  А что произойдет, если разделить $\mathbf{x^2}$ на $\mathbf{x^5}$?

  $$ \frac{x^2}{x^5} = \frac{x \times x}{x \times x \times x \times x \times x} = \frac{1}{x \times x \times x} = \frac{1}{x^3} = x^{-3}$$


  Как видите, при делении $\mathbf{x^2}$ на $\mathbf{x^5}$ можно сократить два $\mathbf{x}$ и в числителе, и в знаменателе, в результате чего получаем $\mathbf{1/x^3}$. Если один и тот же множитель встречается и вверху, и внизу дроби, его можно сократить.

  Чем же является $\mathbf{x^{-3}}$? Отрицательный показатель степени это ещё один способ записать операцию возведения в степень, когда основание находится в знаменателе. Поэтому $\mathbf{x^{-3}}$ эквивалентно $\mathbf{1/x^3}$.

  Снова обратим внимание на правило произведения: его можно применять и к отрицательным степеням. Поэтому деление чисел в степени можно представить как умножение. В нашем примере мы можем превратить показатель "$\mathbf{5}$" у $\mathbf{x^5}$ в отрицательный с последующим его умножением на $x^2$:

  $$
  \frac{x^2}{x^5} = x^2 \times \frac{1}{x^5} = x^2 x^{-5} = x^{2+(-5)} = x^{-3}
  $$  

  В заключение давайте поймем почему любое число в степени 0 равно 1?

  $$
  x^0 = 1
  $$


  Для понимая этого давайте примем, что любое число разделённое на само себя, даёт 1. Когда у вас есть выражение $\mathbf{x^3 / x^3}$, то исходя из этого принципа  оно упрощается до 1. Но упростить это выражение можно и иначе:

  $$ 1 = \frac{x^3}{x^3} = x^3 \times x^{-3} = x^{3 + (- 3)} = x^0 $$

  Если немного поумничать 🧠, то исходя из свойства транзитивности, которое гласит: если $a = b$ и $b = c$, следовательно, $a = c$, получаем, что $x^0 = 1$.

</details>

* Дробные показатели степени являются альтернативным способом представления корней. Например, $\sqrt{4} = 4^{1/2} = 2$ и $\sqrt[3]{8} = 8^{1/3} = 2$

* Иррациональные числа могут использоваться как показатели степени. Hапример, $8^\pi \approx 687.2913$

<details>
  <summary>Текст</summary>

  А что насчёт дробных показателей степени? Они являются альтернативным способом представления корней,
  например, квадратного корня. Для краткого напоминания, вопрос о корне второй степени звучит как "Какое число, умноженное само на себя, даст мне 4?", что, конечно, равно 2. Обратите внимание, что $4^{1/2}$   — это то же самое, что и $\sqrt 4$:

  $$ 4^{1/2} = \sqrt 4 = 2 $$

  Кубические корни похожи на квадратные, однако они определяют число, которое, будучи умноженным на себя трижды, даёт число под корнем. Кубический корень из 8 обозначается как $\sqrt[3] 8$ и отвечает на вопрос: "Какое число, умноженное на себя трижды, даст 8?" Ответом будет 2, ведь $2 \times 2 \times 2 = 8$. В форме степени кубический корень также может быть представлен через дробную степень, так что $\sqrt[3] 8$ может быть записан и как $8^{1/3}$:

  $$ 8^{1/3} = \sqrt[3] 8 = 2 $$  

  Что произойдет при трехкратном умножении кубического корня из 8 на самого себя? Такая операция должна привести к исчезновению кубического корня, возвращая нам исходное число 8, находящиеся под корнем. В соответствии с этими рассужадениями мы можем проверить как сработает степенное представление кубического корня. Так произведение эквивалентно суммированию степеней, поэтому три раза суммируя 1/3 мы получаем 1, а 8 в 1 степени есть просто 8. Поэтому мы снова убедились, что степенное представление корней работает корректно.

  $\sqrt[3] 8  \sqrt[3] 8  \sqrt[3] 8 = 8^{1/3} 8^{1/3} 8^{1/3} = 8^{1/3 + 1/3 + 1/3} = 8^{1} = 8$

  Еще одно ключевое свойство степеней: когда число в степени возводится в степень, показатели степеней перемножаются. Следовательно, выражение $(8^3)^2$ упрощается до $8^6$:

  $$ (8^3)^2 = 8^{3 \times 2} = 8^6 $$

  Если вас интересует, почему так происходит, раскройте выражение и примените правило сложения:

  $$ (8^3)^2 = 8^3 * 8^3 = 8^{3+3} = 8^6$$


  И последнее, что следует сказать о степенях. Это как нужно понимать дробные показатели степени, у которых в числителе число, отличное от 1, как, например, $8^{\frac{2}{3}}$? Это означает, что сначала вычисляется кубический корень из 8, а затем результат возводится в квадрат. Обратите на это внимание:

  $$ 8^{\frac{2}{3}} = (8^{\frac{1}{3}})^2 = 2^2 = 4 $$

  Как это неудивительно, но в качестве показателей степени могут использоваться иррациональные числа, например $8^\pi$ приблизительно равно 687.2913. Этот факт может казаться на первый взгляд сложным для понимания, и неудивительно! Однако, по сути, можно вычислить число в иррациональной степени, приближая его рациональным числом. Это как раз то, что происходит в компьютерных вычислениях, ведь компьютеры способны оперировать лишь ограниченным количеством цифр после запятой. К примеру, число $\pi$ имеет бесконечный ряд десятичных знаков после запятой. Однако, используя первые 11 цифр, 3.1415926535, мы можем приблизительно представить $\pi$ в виде рационального числа (31415926535 / 10000000000). Это в свою очередь позволяет нам приближенно возвесит 8 в степень pi.

  $$ 8^{\pi} \approx 8^{\frac{31415926535}{10000000000}} \approx 687.2913 $$
</details>

# Логарифмы

* **Логарифм** — это функция, определяющая степень, в которую нужно возвести основание, чтобы получить заданное число. Например, в случае  $\mathbf{2^x = 8}$, логарифмическая форма записи будет $\mathbf{\log_2 8 = x}$, что означает $\mathbf{x = 3}$. Это позволяет выделить переменную $\mathbf{x}$ в изолированной форме

* **Логарифмы** широко применяются в различных областях, таких как оценка силы землетрясений, регулировка звука в аудиосистемах, анализ данных и машинное обучение

<details>
  <summary>Текст</summary>

  **Логарифм — это математическая функция, позволяющая определить, в какую степень нужно возвести основание, чтобы получить заданное число**. На первый взгляд, такая функция выглядит абстрактной, но на деле логарифм имеет огромное количество практических применений. Будь то оценка силы землетрясений или регулировка звука в аудиосистеме, логарифмы окружают нас повсюду. Особенно важную роль они играют в сфере машинного обучения и анализа данных. В частности, логарифмы являются ключевым элементом логистической регрессии. Проще всего начать понимание логарифмов с вопроса: "В какую степень необходимо возвести $2$, чтобы получить $8$?". Математически это можно выразить следующим образом:

  $$
  2^x = 8
  $$

  Мы интуитивно знаем ответ, $x = 3$, но нам нужен более понятный способ выразить это математическое сообщение. Для этого и предназначена функция логарифма log().

  $$ log_2 8 = x $$

  Исходя из представленного выше логарифмического выражения, наше основание равно 2, и мы ищем такую степень, чтобы получить результат 8. Переменный показатель степени можно выразить через логарифм следующим образом:

  $$a^x = b$$

  $$x = log_a b$$

  Это метод выделения переменной $x$  в изолированном виде. *Пример 12* демонстрирует методику расчёта данного логарифма в Python.
</details>

```python
# Пример 12: Расчет логарифма.
from math import log, e
# 2 в какой степени даст мне 8?
x = log(e)
print(x)
```

## Натуральный логарифм

  * Использование числа Эйлера $\mathbf{e}$ как основания превращает логарифм в натуральный, который обозначается как $\mathbf{\ln(10)}$ или $ \mathbf{\log_e(10)}$, и в Python для этого используется функция $\mathbf{\text{log()}}$ с основанием $\mathbf{e}$ по умолчанию

<details>
  <summary>Текст</summary>

  Использование числа Эйлера $e$ как основания превращает логарифм в **натуральный**. В зависимости от программной среды, для обозначения натурального логарифма может применяться функция $\mathbf{ln()}$ вместо $\mathbf{log()}$. Так, запись натурального логарифма числа $\mathbf{10}$, вычисляющего степень для $e$, чтобы получить $\mathbf{10}$, упрощается до $\mathbf{ln(10)}$ вместо $\mathbf{\log_{e}{10}}$:

  $$
  \log_{e}{10} = \ln{10}
  $$

  Тем не менее в **Python** для обозначения натурального логарифма используется функция $\mathbf{log()}$. По умолчанию основанием для этой функции служит $\mathbf{e}$, если не указан второй аргумент, задающий основание.
</details>

```python
# Пример 15: Натуральный логарифм
from math import log
# e возведенное в какую степень даёт 10?
x = log(10)
print(x)
```

Рисунок 2. Функция которая приближается к 0, но никогда сама не становится нулем

<details>
  <summary>Текст</summary>
  
  Мы рассматриваем только положительные значения $x$. Обратите внимание, что по мере бесконечного увеличения $x$, $f(x)$ приближается к 0. Удивительно, но $f(x)$ никогда на самом деле не достигает 0. Она просто бесконечно продолжает приближаться к нему.

  Следовательно, характер поведения данной функции определён так: с  увеличением $x$ до бесконечности она асимптотически приближается к нулю, однако никогда фактически нуля не достигает. Математически это выражается через понятие предела:

  $$ \lim_{x \to \infty} \frac{1}{x} = 0 $$

  Этот принцип "бесконечного приближения, но недостижения" встречается достаточно часто, в частности в изучении производных и интегралов.

  При помощи SymPy можно вычислить предел функции $f(x) = \frac{1}{x}$, когда $x$ стремится к бесконечности $\infty$, как показано в Примере 1-16. Следует отметить, что в SymPy символ бесконечности $\infty$ обозначается как oo.

</details>

```python
# Пример 16: Расчет пределов с помощью sympy
from sympy import *
x = symbols('x')
f = 1 / x
result = limit(f, x, oo)
print(result) # 0
```

<details>
  <summary>Текст</summary>

  **Число Эйлера** $e$ может быть тоже выраженно через предел:

  $$ \lim_{n \to \infty} \left(1 + \frac{1}{n}\right)^n = e = 2.71828169413... $$

  Забавно, что когда мы вычисляем число Эйлера с помощью пределов в SymPy, SymPy сразу же распознает его как число Эйлера. Мы можем вызвать функцию evalf(), чтобы отобразить $e$ как число:

</details>

```python
# Пример 17: Расчет числа е через предел
from sympy import *
n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)
print(result) # E
print(result.evalf()) # 2.71828182845905
```
