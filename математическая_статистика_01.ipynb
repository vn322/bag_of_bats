# Статистика

Хотя **статистика** не всегда получает заслуженное признание, она является основой многих инноваций, основанных на данных. **Машинное обучение**, например, является статистическим инструментом, который ищет возможные гипотезы для установления связей между различными переменными в данных. # Описательная и процедурная статистики

Что приходит вам на ум, когда вы слышите слово **«статистика»?** Возможно, это вычисление среднего значения, медианы, моды, построение графиков, кривые Гаусса и другие инструменты для описания данных? Это самая распространенная и понятная часть статистики — **описательная статистика**, которую мы используем для обобщения и представления данных. В конце концов, что проще: пролистывать миллион записей в данных или получить их обобщение? Мы начнем с изучения именно этой области статистики.

**Индуктивная статистика**, с другой стороны, пытается выявить характеристики более широкой популяции, часто на основе выборки. Эта часть статистики часто неправильно понимается и менее интуитивна, чем описательная. Часто нас интересует изучение группы, которая слишком велика для наблюдения (например, средний рост подростков в Европе), и нам приходится полагаться на небольшую **выборку** представителей этой группы, чтобы сделать выводы. Как вы можете догадаться, это не так просто, ведь мы пытаемся представить всю **популяцию** с помощью **выборки**, которая может не быть репрезентативной. ## Смещение (Bias)

Если мы собираемся делать выводы о характеристиках популяции на основе **выборки**, важно, чтобы эта выборка была как можно более случайной, иначе наши выводы могут быть искажены. Вот пример. Допустим, я студент МГУ. Я хочу выяснить среднее количество часов, которое студенты университетов в России проводят на YouTube в неделю. Я выхожу за пределы своего общежития и начинаю опрашивать случайных студентов, которые проходят мимо, завершая сбор данных за несколько часов. В чем проблема?

Проблема в том, что наша выборка студентов будет **смещенной**, то есть она исказит наши результаты, избыточно представляя одну группу за счет других. Мое исследование определило популяцию как «студенты университетов в России», а не «студенты МГУ». Я опрашиваю студентов только одного конкретного университета, чтобы представить всех студентов университетов по всей России! Это действительно справедливо?

Разве использование их для представления всей страны не исказит результаты? Возможно, это связано с особенностями климата и жизни в большом городе. Поэтому просмотр видеороликов на YouTube может быть распространенным времяпрепровождением. Однако **смещение** не всегда географическое. Допустим, я прилагаю искренние усилия, чтобы опросить студентов по всей России. Что, если студенты, которые активно пользуются социальными сетями и видят опрос, также склонны больше смотреть YouTube? Если они часто находятся в социальных сетях, вероятно, они не против проводить больше времени перед экраном. Легко представить, что у них всегда под рукой YouTube и другие стриминговые сервисы. Этот конкретный тип **смещения**, когда определенная группа чаще включается в выборку, называется смещением самовыбора.

Кажется, победить это невозможно, правда? Если задуматься, **смещение** данных кажется неизбежным! И часто это действительно так. Эта проблема **смещения** данных дорогостоящая и трудная для преодоления, и машинное обучение особенно уязвимо к ней.

# Среднее значение и взвешенное среднее значение.

**Среднее значение** – это среднее арифметическое набора значений. Операция проста: сложите все значения и разделите на их количество. **Среднее значение** полезно, потому что показывает «центр тяжести» набора наблюдаемых значений.

Среднее значение вычисляется одинаково как для популяций, так и для выборок. *Пример 1* показывает выборку из восьми значений и как вычислить их среднее значение на Python.

```python
# Пример 1: Вычисление среднего значения на Python
# Количество домашних животных у каждого человека
sample = [1, 3, 2, 5, 7, 0, 2, 3]
mean = sum(sample) / len(sample)
print(mean) # выводит 2.875
```

Существует две версии среднего значения, которые вы увидите: среднее значение выборки $ \bar{x} $ и среднее значение популяции $ \mu $, как показано здесь:

$$
\bar{x} = \frac{x_1 + x_2 + x_3 + \ldots + x_n}{n} = \frac{\Sigma x_i}{n}
$$

$$
 \mu = \frac{x_1 + x_2 + x_3 + \ldots + x_n}{N} = \frac{\Sigma x_i}{N}
 $$

Помните, что символ суммирования $ \Sigma \ $ означает сложение всех элементов. $ n $ и $ N $ представляют размер выборки и популяции соответственно, но математически они обозначают одно и то же: количество элементов. То же самое относится к выборочному среднему $ \bar{x} $ (называемому «x-c подчеркиванием») и среднему значению популяции $ \mu $ («мю»). Оба $ \bar{x} $ и $ \mu $ это одно и то же вычисление, просто разные названия в зависимости от того, работаем ли мы с выборкой или популяцией.

**Среднее значение**, вам знакомо, но вот что мало кто знает: **среднее значение** на самом деле является **взвешенным средним**, называемым взвешенным средним значением. Обычное среднее значение, которое мы используем, придает равное значение каждому элементу. Но мы можем изменить **среднее значение** и задать каждому элементу разный вес:

$$
\text{ВЗВЕШЕННОЕ СРЕДНЕЕ} = \frac{x_1 \cdot w_1 + x_2 \cdot w_2 + x_3 \cdot w_3 + \ldots + x_n \cdot w_n}{w_1 + w_2 + w_3 + \ldots + w_n}
$$

Это может быть полезно, когда мы хотим, чтобы некоторые значения оказывали большее влияние на **среднее значение,** чем другие. Распространённый пример этого – взвешивание академических экзаменов для получения итоговой оценки. Если у вас есть три промежуточных экзамена и один финальный экзамен, и мы присваиваем каждому из трех промежуточных экзаменов вес **20%**, а **финальному экзамену** – вес **40%** от итоговой оценки, это можно выразить, как показано в *примере 2*.




```python
# Пример 2: Вычисление взвешенного среднего на Python
# Три экзамена с весом по 0.20 каждый и финальный экзамен с весом 0.40
sample = [90, 80, 63, 87]
weights = [.20, .20, .20, .40]
weighted_mean = sum(s * w for s, w in zip(sample, weights)) / sum(weights)
print(weighted_mean) # выводит 81.4
```

Мы умножаем каждую оценку экзамена на соответствующий вес. Вместо деления на количество оценок, мы делим на сумму весов. Веса не обязательно должны быть в процентах, поскольку любые числа, используемые для весов, будут пропорциональны. В *примере 3* мы присваиваем каждому из трех промежуточных экзаменов вес **"1"**, а финальному экзамену вес **"2"**, что дает ему вдвое больший вес. Мы все равно получим тот же результат **81.4**, поскольку эти значения останутся пропорциональными.



```python
# Пример 3: Вычисление взвешенного среднего на Python
# Три экзамена с весом по 1.0 каждый и финальный экзамен с весом 2.0
sample = [90, 80, 63, 87]
weights = [1.0, 1.0, 1.0, 2.0]
weighted_mean = sum(s * w for s, w in zip(sample, weights)) / sum(weights)
print(weighted_mean) # выводит 81.4
```

# Медиана
**Медиана** – это серединное значение в наборе упорядоченных значений. Вы упорядочиваете значения по порядку, и медианой будет центральное значение. Если количество значений четное, медиана определяется как среднее значение двух центральных значений. В *примере 4* медиана количества домашних животных в нашей выборке является **7**:

$$0, 1, 5, *7*, 9, 10, 14$$

```python
# Примере 4: Нахождение медианы
sample = [0, 1, 5, 7, 9, 10, 14]

def median(values):
    ordered = sorted(values)
    n = len(ordered)
    mid = n // 2
    if n % 2 == 0:
        return (ordered[mid - 1] + ordered[mid]) / 2.0
    else:
        return ordered[mid]

print(median(sample)) # выводит 7
```

Медиана может быть полезной альтернативой среднему значению, когда данные искажены выбросами — значениями, которые значительно больше или меньше по сравнению с остальными значениями. Вот интересная история, чтобы понять, почему это так.

В **1986** году средняя начальная годовая зарплата выпускников географического факультета Университета Северной Каролины в Чапел-Хилл составляла **250 000** долларов. В других университетах эта средняя цифра составляла **22 000** долларов. Ух ты, программа по географии в Чапел-Хилл должна быть потрясающей!

Но что на самом деле было таким доходным в программе по географии Чапел-Хилла? Ну… одним из их выпускников был **Майкл Джордан**. Один из самых известных игроков НБА всех времён действительно получил диплом по географии в Чапел-Хилле. Однако он начал свою карьеру, играя в баскетбол, а не изучая карты. Очевидно, что это выброс — значение, которое сильно искажает среднюю.

Поэтому **медиана может быть предпочтительнее в ситуациях, когда данные содержат множество выбросов** (например, в данных о доходах). Медиана менее чувствительна к выбросам и разделяет данные строго пополам на основе их порядка, а не их точного значения на числовой линии. Когда ваша медиана сильно отличается от среднего значения, это указывает на наличие выборки данных с выбросами и смещениями.

# Мода
**Мода** — это наиболее часто встречающееся значение или набор значений. Она становится особенно полезной, когда ваши данные повторяются, и вы хотите определить, какие значения встречаются наиболее часто.

Когда ни одно значение не встречается больше одного раза, **мода** отсутствует. Если два значения встречаются с одинаковой частотой, тогда набор данных считается бимодальным. В *примере 5* мы вычисляем моду для нашего набора данных о домашних животных, и действительно, мы видим, что этот набор данных бимодален, так как и 2, и 3 встречаются одинаково часто.

```python
# Пример 5: Вычисление моды на Python
# Количество домашних животных, которых имеет каждый человек
from collections import defaultdict

sample = [1, 3, 2, 5, 7, 0, 2, 3]

def mode(values):
    counts = defaultdict(lambda: 0)
    for s in values:
        counts[s] += 1
    max_count = max(counts.values())
    modes = [v for v in set(values) if counts[v] == max_count]
    return modes

print(mode(sample)) # [2, 3]
```

На практике мода используется не так часто, если только ваши данные не имеют тенденции к повторяемости. Это чаще всего встречается с целыми числами, категориями и другими дискретными переменными.

# Дисперсия и стандартное отклонение
Одна вещь, которая может сбивать людей с толку при расчете **дисперсии** и **стандартного отклонения**, — это различия в расчетах для выборки и для всей совокупности. Постараемся изложить эти различия как можно более ясно.

### Дисперсия и стандартное отклонение для всей совокупности.
При описании данных нас часто интересует измерение различий между **средним значением** и каждой отдельной записью. Это дает нам представление о том, насколько «разбросаны» данные.

Предположим, я хочу изучить количество домашних животных, которых имеют участники моей рабочей команды (обратите внимание, что я определяю это как всю мою популяцию, а не выборку). У меня в команде семь человек.

Я вычисляю среднее количество домашних животных, которых имеют все участники команды, и получаю **6.571**. Давайте вычтем это среднее из каждого значения. Это покажет нам, насколько каждое значение отличается от среднего, как показано в *Таблице 1*.

| Значение | Среднее | Разница |
|----------|---------|---------|
| 0        | 6.57    | -6.57   |
| 1        | 6.57    | -5.57   |
| 5        | 6.57    | -1.57   |
| 7        | 6.57    | 0.43    |
| 9        | 6.57    | 2.43    |
| 10       | 6.57    | 3.43    |

Таблица 1. Отклонение от среднего количества числа животных


Теперь подумайте, почему эта информация может быть полезной. Различия показывают нам, насколько разбросаны данные и насколько значения отклоняются от среднего. Существует ли способ свести эти различия к одному числу, чтобы быстро описать **степень разброса** данных?

Вы можете подумать, что стоит взять среднее этих различий, но отрицательные и положительные значения будут компенсировать друг друга при суммировании. Мы могли бы суммировать абсолютные значения (избавиться от знаков минуса и сделать все значения положительными). Еще лучший подход — возводить эти различия в квадрат перед суммированием. Это не только устраняет отрицательные значения (так как квадрат отрицательного числа дает положительное значение), но и усиливает большие различия, и с ними математически легче работать. После этого нужно усреднить возведенные в квадрат различия. Это даст нам **дисперсию**, меру того, насколько разбросаны наши данные.

Вот математическая формула, показывающая, как рассчитывать **дисперсию**:

$$
\text{ДИСПЕРСИЯ В ГЕНЕРАЛЬНОЙ СОВОКУПНОСТИ} =
\frac{(x_1 - \text{среднее})^2 + (x_2 - \text{среднее})^2 + \ldots + (x_n - \text{среднее})^2}{N}
$$

Более кратко, **дисперсия** для **генеральной совокупности** выглядит так:

$$
\sigma^2 = \frac{\sum (x_i - \mu)^2}{N}
$$

Расчёт дисперсии **генеральной совокупности** на примере данных о домашних животных в **Python** показан в *примере 6*.



```python
# Пример 6: Вычисление дисперсии в Python
# Количество домашних животных у каждого человека
data = [0, 1, 5, 7, 9, 10, 14]

def variance(values):
    mean = sum(values) / len(values)
    _variance = sum((v - mean) ** 2 for v in values) / len(values)
    return _variance

print(variance(data)) # выводит 21.387755102040813
```

Итак, дисперсия количества домашних животных у сотрудников моей фирмы равна **21.3877**. Хорошо, но что это означает на самом деле? Логично заключить, что более высокая дисперсия означает больший разброс, но как соотнести это значение с нашими данными? Это число больше любого из наших наблюдений, потому что мы возводили в квадрат и суммировали значения, что перевело его на совершенно другую шкалу. Как мы можем вернуть его обратно к нашей оригинальной шкале?

Противоположное действие к возведению в квадрат — это извлечение квадратного корня, так что давайте возьмём квадратный корень из дисперсии, что даст нам **стандартное отклонение**. Это **дисперсия**, приведённая к числу, выраженному в «количестве домашних животных», что делает её немного более осмысленной:

$$
\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}}
$$

В **Python** мы можем повторно использовать функцию variance() и применить sqrt() к её результату. Теперь у нас есть функция std_dev(), представленная в *примере 7*.



```python
# Пример 7: Вычисление стандартного отклонения в Python
from math import sqrt

# Количество домашних животных у каждого человека
data = [0, 1, 5, 7, 9, 10, 14]

def variance(values):
    mean = sum(values) / len(values)
    _variance = sum((v - mean) ** 2 for v in values) / len(values)
    return _variance

def std_dev(values):
    return sqrt(variance(values))

print(std_dev(data)) # выводит 4.623353
```

При запуске кода в *примере 7*, вы увидите, что наше стандартное отклонение примерно равно **4.62** домашних животных. Таким образом, мы можем выразить нашу разбросанность на той же шкале, с которой мы начали, и это делает нашу дисперсию немного легче для интерпретации.

#Дисперсия выборки и стандартное отклонение

В предыдущем разделе мы говорили о дисперсии и стандартном отклонении для генеральной совокупности.
Однако, существует важное изменение, которое необходимо применить к этим двум формулам, когда мы вычисляем показатели для выборки:

$$
s^2 = \frac{\sum (x_i - \bar{x})^2}{n - 1}
$$

$$
s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n - 1}}
$$

Вы заметили разницу? Когда мы усредняем квадраты отклонений, мы делим не на $n$, а на $n–1$. Почему мы так делаем? Это необходимо для уменьшения **смещения** в **выборке** и избежания недооценки дисперсии совокупности на основе нашей выборки. Исключая один элемент из знаменателя, мы увеличиваем дисперсию, что позволяет более точно учитывать неопределенность в нашей выборке.

Если наши данные о домашних животных являются **выборкой**, а не **генеральной совокупностью**, мы должны внести соответствующие корректировки. В *примере 8* мы модифицируем свой предыдущий код Python для функций `variance()` и `std_dev()`, добавляя параметр `is_sample`, который, если установлен в `True`, уменьшит знаменатель на 1 при вычислении **дисперсии**.

Обратите внимание, что в *примере 8* значения дисперсии и стандартного отклонения увеличились по сравнению с предыдущими примерами, в которых они рассчитывались как показатели генеральной совокупности. В *примере 7* стандартное отклонение было около **4.62**, при условии, что это генеральная совокупность. Но здесь, рассматривая данные как выборку (уменьшив знаменатель дисперсии на 1), мы получаем примерно **4.99**. Это правильно, так как выборка может быть смещенной и неидеально представлять генеральную совокупность. Следовательно, мы увеличиваем дисперсию (а значит и стандартное отклонение), чтобы повысить нашу оценку разброса значений. Большая **дисперсия/стандартное** отклонение показывает меньше уверенности с большим диапазоном.

Так же, как для среднего значения ($\bar{x}$ для выборки и $\mu$ для совокупности), часто используются определенные символы для дисперсии и стандартного отклонения. Стандартное отклонение для выборки и совокупности обозначаются $s$ и $\sigma$ соответственно. Вот формулы стандартного отклонения для выборки и совокупности:

$$
s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n - 1}}
$$

$$
\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}}
$$

**Дисперсия** будет квадратом этих двух формул, отменяя извлечение квадратного корня. Следовательно, **дисперсия** для **выборки** и **совокупности** обозначается $s^2$ и $\sigma^2$ соответственно:

$$
s^2 = \frac{\sum (x_i - \bar{x})^2}{n - 1}
$$

$$
\sigma^2 = \frac{\sum (x_i - \mu)^2}{N}
$$

Опять же, квадрат в написании означает, что для получения **стандартного отклонения** необходимо извлечь квадратный корень.


```python
### Пример 8: Вычисление стандартного отклонения для выборки
from math import sqrt

# Количество домашних животных у каждого человека
data = [0, 1, 5, 7, 9, 10, 14]

def variance(values, is_sample: bool = False):
    mean = sum(values) / len(values)
    _variance = sum((v - mean) ** 2 for v in values) / (len(values) - (1 if is_sample else 0))
    return _variance

def std_dev(values, is_sample: bool = False):
    return sqrt(variance(values, is_sample))

print("VARIANCE = {}".format(variance(data, is_sample=True))) # 24.95238095238095
print("STD DEV = {}".format(std_dev(data, is_sample=True)))   # 4.99523582550223
```

### Свойства нормального распределения

**Нормальное распределение** имеет несколько важных свойств, которые делают его полезным:

- Оно симметрично; обе стороны идентично отражены относительно среднего значения, которое находится в центре.
- Большая часть массы сосредоточена в центре вокруг среднего значения.
- Оно имеет разброс (может быть узким или широким), который определяется стандартным отклонением.
- “Хвосты” представляют наименее вероятные результаты и бесконечно приближаются к нулю, но никогда не достигают его.
- Оно напоминает множество явлений в природе и повседневной жизни, и даже обобщает ненормальные проблемы благодаря центральной предельной теореме, о которой мы поговорим чуть позже.

### Плотность вероятности

Стандартное отклонение играет важную роль в нормальном распределении, так как оно определяет, насколько "разбросанным" оно является. Оно фактически является одним из параметров наряду со средним значением. Функция плотности вероятности, которая формирует нормальное распределение, выглядит следующим образом:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left(\frac{x - \mu}{\sigma}\right)^2}
$$




```python
# Пример 9
import math

def normal_pdf(x, mean, std_dev):
    return (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(-0.5 * ((x - mean) / std_dev) ** 2)
```

Здесь много всего, что требуется разобрать в этой формуле, но важно то, что она принимает среднее значение и стандартное отклонение в качестве параметров, а также значение $x$, чтобы вы могли определить вероятность для данного значения.

Так же как бета-распределение, нормальное распределение является непрерывным. Это означает, что для получения вероятности нам нужно интегрировать диапазон значений $x$, чтобы найти площадь.

На практике же мы будем использовать SciPy для выполнения этих расчетов.

# Функция кумулятивной плотности

В случае нормального распределения вертикальная ось не отображает вероятность, а скорее вероятность для данных. Чтобы найти вероятность, нам нужно рассмотреть заданный диапазон и найти площадь под кривой для этого диапазона. Предположим, я хочу найти вероятность того, что вес золотистого ретривера находится в пределах от **27** до **31** килограмм. *Рисунок 4* показывает диапазон, для которого мы хотим найти площадь.



```python
#@title <b> Рисунок </b>

import numpy as np
import matplotlib.pyplot as plt

# Устанавливаем параметры нормального распределения
mean_value = 30    # Среднее значение
std_dev_value = 5  # Стандартное отклонение
num_samples = 1000 # Количество сэмплов

# Генерируем случайные веса по нормальному распределению
weights = np.random.normal(mean_value, std_dev_value, num_samples)

# Построение графика
plt.figure(figsize=(10, 6))

# График плотности вероятности
plt.xlim([10, 50])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-((x - mean_value) ** 2) / (2 * std_dev_value ** 2)) / (np.sqrt(2 * np.pi) * std_dev_value)
plt.plot(x, p, 'k', linewidth=2, label='Нормальное распределение PDF')

# Закрасить под кривой от 27 до 31
x_fill = np.linspace(27, 31, 100)
p_fill = np.exp(-((x_fill - mean_value) ** 2) / (2 * std_dev_value ** 2)) / (np.sqrt(2 * np.pi) * std_dev_value)
plt.fill_between(x_fill, p_fill, color='red', alpha=0.3, label='Интервал от 27 до 31')

# Наименование осей и заголовок
plt.title('Нормальное распределение весов золотистых ретриверов')
plt.xlabel('Вес (кг)')
plt.ylabel('Плотность вероятности')

#plt.legend()
plt.grid(True)
plt.show()
```

Рисунок 4. CDF расчет вероятности, что лабородор будет в диапазоне 27 и 31 кг

Мы уже решали эту задачу с бета-распределением, и, как и в случае с бета-распределением, существует **функция кумулятивной плотности** (*англ* Сommulative distribution funciton - CDF). Давайте воспользуемся этим подходом. Как мы узнали в предыдущей лекции, CDF выдает  площадь до заданного значения x для заданного распределения. Давайте посмотрим, как выглядит CDF для нашего нормального распределения весов золотистых ретриверов, и сравним его с распределение плотности вероятности (PDF) на *рисунке 5*.

Обратите внимание на взаимосвязь между двумя графиками. CDF, который представляет собой S-образную кривую (называемую сигмоидой), показывает площадь до данного диапазона в PDF. На *рисунке 6* видно, что если мы захватываем площадь от минус бесконечности до 30.00 (среднего значения), наша CDF показывает значение ровно 0.5, или 50%! Эта площадь в 0.5 или 50% до среднего значения легко определяется благодаря **симметрии** нашего нормального распределения, и мы можем ожидать, что другая сторона кривой **нормального распределения** также будет занимать 50% площади.

Для вычисления этой площади до значения 30.30 в Python с использованием SciPy, используйте функцию `norm.cdf()`, как показано в Примере 3-10.






```python
# Пример 10. Накопленная функция распределения (CDF) для нормального распределения в Python
from scipy.stats import norm

mean = 29.22
std_dev = 1.36
x = norm.cdf(29.22, mean, std_dev)
print(x)  # выводит 0.5
```

Так же, как мы делали в предыдущей лекции, мы можем вычислить площадь для среднего диапазона вычитанием областей. Если мы хотим найти вероятность того, что вес золотистого ретривера находится в пределах от 28 до 30 килограмм, мы вычислим площадь до 30 и вычтем площадь до 28, как показано на *рисунке 7*. Сделать это в **Python** с использованием **SciPy** просто - вычтите два значения CDF, как показано в *примере 11*.




```python
# Пример 11. Получение вероятности для среднего диапазона с использованием CDF
from scipy.stats import norm
mean = 30.00
std_dev = 5.00
x = norm.cdf(30, mean, std_dev) - norm.cdf(28, mean, std_dev)
print(x)  # выводит 0.4920450147062894
```

Вы обнаружите, что вероятность того, что вес золотистого ретривера находится в пределах от 28 до 30 килограмм, составляет **0.155**, или примерно **15.5%**.

```python
#@title <b> Рисунок </b>

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Устанавливаем параметры нормального распределения
mean_value = 30    # Среднее значение
std_dev_value = 5  # Стандартное отклонение
num_samples = 1000 # Количество сэмплов

# Генерируем случайные веса по нормальному распределению
weights = np.random.normal(mean_value, std_dev_value, num_samples)

# Построение графика
plt.figure(figsize=(10, 6))

# График плотности вероятности (PDF)
plt.xlim([10, 50])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-((x - mean_value) ** 2) / (2 * std_dev_value ** 2)) / (np.sqrt(2 * np.pi) * std_dev_value)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Построение накопленной функции распределения (CDF)
cdf = stats.norm.cdf(x, mean_value, std_dev_value)
plt.plot(x, cdf, 'b', linewidth=2, label='CDF')

# Наименование осей и заголовок
plt.title('Нормальное распределение весов золотистых ретриверов')
plt.xlabel('Вес (кг)')
plt.ylabel('Плотность вероятности / CDF')
plt.legend()
plt.grid(True)
plt.show()
```

# Обратная функция кумулятивной плотности (Inverse CDF)

Когда мы будем проводить тестирование гипотез в следующей лекции, мы встретим ситуации, где нам нужно будет найти значение на графике CDF и вернуть соответствующее значение $x$. Конечно, это обратное использование CDF, поэтому мы будем использовать обратную CDF, которая меняет оси, как показано на *рисунке 8*.

Таким образом, мы можем найти вероятность и вернуть соответствующее значение $x$, и в **SciPy** для этого мы будем использовать функцию `norm.ppf()`. Например, я хочу найти вес, под который попадает **95%** золотистых ретриверов. Это легко сделать с использованием обратной CDF в *примере 12*.



```python
# Пример 12: Использование обратной CDF (называемой ppf()) в Python
from scipy.stats import norm

x = norm.ppf(0.95, loc=30.0, scale=5.0)
print(x)
```


Выясняется, что **95%** золотистых ретриверов весят **38.22** килограмма или меньше.

Вы также можете использовать обратную CDF для генерации случайных чисел, следующих нормальному распределению. Если я хочу создать симуляцию, которая генерирует тысячу реалистичных весов золотистых ретриверов, я просто генерирую случайное значение в диапазоне от **0.0** до **1.0**, передаю его в обратную CDF и возвращаю значение веса, как показано в *примере 13*.

```python
# Пример 13: Генерация случайных весов ретриверов из нормального распределения
import random
from scipy.stats import norm

for i in range(0, 10):
    random_p = random.uniform(0.0, 1.0)
    random_weight = norm.ppf(random_p, loc=29.22, scale=1.36)
    print(random_weight)
```

Конечно, **NumPy** и другие библиотеки могут генерировать случайные значения из распределения за вас, но этот пример показывает один из случаев использования, когда обратная **CDF** полезна.

```python
#@title <b> Рисунок </b>

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Устанавливаем параметры нормального распределения
mean_value = 30    # Среднее значение
std_dev_value = 5  # Стандартное отклонение
num_samples = 1000 # Количество сэмплов

# Генерируем случайные веса по нормальному распределению
weights = np.random.normal(mean_value, std_dev_value, num_samples)

# Построение графика
plt.figure(figsize=(10, 6))

# Создаем массив значений вероятности (от 0 до 1), который будем использовать для построения обратной CDF
probabilities = np.linspace(0.0001, 0.9999, 10000)  # используем значения от 0.01 до 0.99, чтобы избежать ошибки
inverse_cdf = stats.norm.ppf(probabilities, mean_value, std_dev_value)

# График обратной накопленной функции распределения (Inverse CDF или PPF)
plt.plot(probabilities, inverse_cdf, 'b', linewidth=2, label='Inverse CDF')
plt.ylim([5, 55])

# Наименование осей и заголовок
plt.title('Обратная накопленная функция распределения (Inverse CDF) весов золотистых ретриверов')
plt.xlabel('Вероятность')
plt.ylabel('Вес (кг)')
plt.legend()
plt.grid(True)
plt.show()
```

### Z-оценки (Z-Scores)

Часто **нормальное распределение** масштабируется так, чтобы среднее было 0, а стандартное отклонение - 1. Это называется стандартным нормальным распределением. Это позволяет легко сравнивать разброс одного нормального распределения с другим, даже если у них разные средние значения и дисперсии.

Стандартное нормальное распределение выражает все значения $x$ в терминах стандартных отклонений, известными как Z-оценки. Преобразование значения x в Z-оценку использует базовую формулу масштабирования:
$$
z = \frac{x - \mu}{\sigma}
$$

Вот пример. У нас есть два дома из двух разных районов. Район **A** имеет среднюю стоимость дома **10,000,000** рублей и стандартное отклонение **200,000** рублей. Район **B** имеет среднюю стоимость дома **50,000,000** рублей и стандартное отклонение **700,000** рублей.

$$
\mu_A = 10,000,000
$$

$$
\mu_B = 50,000,000
$$

$$
\sigma_A = 200,000
$$

$$
\sigma_B = 700,000
$$

Теперь у нас есть два дома из каждого района. Дом A из района A стоит **9,500,000** рублей, а дом B из района B стоит **52,000,000** рублей. Какой дом дороже относительно среднего значения в своем районе?

$$
x_A = 9,500,000
$$

$$
x_B = 52,000,000
$$

Если мы выразим эти два значения в терминах стандартных отклонений, мы сможем сравнить их относительно среднего значения их района, используя формулу Z-оценки:

$$
z = \frac{x - \mu}{\sigma}
$$

$$
z_A = \frac{9,500,000 - 10,000,000}{200,000} = -2.5
$$

$$
z_B = \frac{52,000,000 - 50,000,000}{700,000} = 2.8571
$$

Таким образом, дом в районе B действительно намного дороже относительно среднего значения домов в своем районе, чем дом в районе A, так как их Z-оценки составляют **-2.5** и **2.8571** соответственно.

Вот как мы можем преобразовать значение $x$ из данного распределения с определённым средним значением и стандартным отклонением в Z-оценку и наоборот, как показано в *примере 14*.


```python
# Пример 14: Преобразование значений в Z-оценки и обратно
def z_score(x, mean, std):
    return (x - mean) / std

def z_to_x(z, mean, std):
    return (z * std) + mean

mean = 10000000
std_dev = 200000
x = 9500000

# Преобразование в Z-оценку и обратно
z = z_score(x, mean, std_dev)
back_to_x = z_to_x(z, mean, std_dev)

print("Z-оценка: {}".format(z))  # выводит Z-оценка: -2.5
print("Обратно к X: {}".format(back_to_x))  # выводит Обратно к X: 9,500,000
```

Функция `z_score()` принимает значение x и масштабирует его в терминах стандартных отклонений, используя среднее значение и стандартное отклонение. Функция `z_to_x()` принимает Z-оценку и преобразует ее обратно в значение $x$. Изучая две функции, вы можете увидеть их алгебраическое соотношение: одна решает для Z-оценки, а другая для значения $x$. Мы преобразуем значение $x$ в **9,500,000** в Z-оценку **-2.5**, а затем преобразуем эту Z-оценку обратно в значение $x$.
